```python
def batch_download(starting_planet_number, ending_planet_number):
    """Function to download all FITS files corresponding to a certain 
    target or series of targets observed across all observational 
    periods. Only works for planets in the format 'Kepler-{id}'"""
    for identifier in range(starting_planet_number, ending_planet_number + 1):
        lpf_collection = search_lightcurvefile("Kepler -" + str(identifier), mission = "Kepler").download_all(download_dir = os.getcwd())
    return "Download complete."
```
