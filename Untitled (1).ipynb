# Downloading and plotting Kepler light curves

### Step 1: Imports and functions

```python
import fnmatch
import os
from astropy.timeseries import TimeSeries
import lightkurve as lk
import pandas as pd
import itertools
import shutil
from lightkurve import search_lightcurvefile

def batch_download(starting_planet, ending_planet, mission_name):
    for identifier in range(int(starting_planet), int(ending_planet) + 1):
        lpf_collection = search_lightcurvefile(str(mission_name) +"-" + str(identifier), mission = mission_name).download_all()
        
def file_pusher(source_dir, target_dir):
    file_names = os.listdir(source_dir)
    for file_name in file_names:
        shutil.move(os.path.join(source_dir, file_name), target_dir)   
```

We begin by taking care of all top level imports. We then write a function, batch_download, that queries the MAST archive for light curve files. To use the batch_download function, pass to it the number of the starting planet, ending planet, and the name of the mission (Kepler or K2 -- TESS naming conventions are different) for which you want to obtain data. The file_pusher function moves the data downloaded by the batch_download function from a source directory, source_dir, to a target directory, target_dir. It automatically deletes the data from the source_dir (by design -- because FITS files are relatively large, I prefer working with them on an external drive, rather than on my laptop's disk.)

### Step 3: Processing the light curve file collections

```python
parent_directory = "/Volumes/ExtremeSSD/exo2/mastDownload/Kepler"
all_times = [[] for x in range(len(fnmatch.filter(os.listdir(parent_directory), 'kplr*')))]
all_fluxes = [[] for x in range(len(fnmatch.filter(os.listdir(parent_directory), 'kplr*')))]
list_of_kepler_ids = []

for subdirectory in os.listdir(parent_directory):
    if fnmatch.fnmatch(subdirectory, "kplr*"):
        list_of_kepler_ids.append(subdirectory[:13])
        jd_times = []
        fluxes = []
        for file in os.listdir(os.path.join(parent_directory, subdirectory)):
            file = os.path.join(parent_directory, subdirectory, file)
            ts = TimeSeries.read(file, format='kepler.fits') 
            times = ts.time.jd
            df = ts.to_pandas()
            pdcsap = (df["pdcsap_flux"])
            lc = lk.LightCurve(times, pdcsap).normalize()
            lc = lc.remove_nans()
            fluxes.append(lc.flux)
            jd_times.append(lc.time)
        jd_times = list(itertools.chain(*jd_times))
        fluxes = list(itertools.chain(*fluxes))
    all_times.append(jd_times)
    all_fluxes.append(fluxes)

all_times = list(all_times for all_times,_ in itertools.groupby(all_times))
all_fluxes = list(all_fluxes for all_fluxes,_ in itertools.groupby(all_fluxes))

for element in all_times:
    if len(element) == 0:
        all_times.remove(element)
        
for element in all_fluxes:
    if len(element) == 0:
        all_fluxes.remove(element)
```

Here is the actual meat and potatoes of this script. To start, we create $n$ blank lists within a list of lists, all_times, where $n = $ the number of targets we specified in batch_download. We also create a blank list of Kepler IDs to be used in case we need to figure out from which file collection some particular values come from. We loop through all subdirectories within the mastDownload/Kepler directory, each of which contains a collection of light curve FITS files. From the names of these subdirectories, we populate the list of Kepler IDs with unique identifiers for each target. We create absolute filepaths for each file within the light curve file collections (i.e. the subdirectories) and use astropy's TimeSeries module to read them in. We convert all times to Julian Days. We then write the entire TimeSeries to a pandas DataFrame. This allows to convert fluxes from some sort of obscure astropy datatype to floats. We then create a lightcurve object out of the time and PDCSAP (cleaned) flux. We normalize it so that we can eventually stitch together lightcurves, and we append to the flux and jd_times lists the normalized flux and the time which corresponds to it. We unravel the list of lists jd_times and fluxes using the itertools.chain method, and append to the ultimate all_times and all_fluxes lists these jd_times and fluxes lists. Obviously, there are as many of these lists as there are file collections. We remove any duplicates that might have been produced by the double loop by using itertools.groupyby, and then remove any blank elements of the final lists that might exist for whatever reason. Let's go ahead and see if this code worked.

```python
for i in range(len(all_times)):
    lc = lk.LightCurve(all_times[i], all_fluxes[i])
    lc.scatter();
```

We plot all the curves, and all the plots appear correct. There are as many plots as there were file collections. There you have it!
